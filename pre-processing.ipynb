```python
import pandas as pd

# Define the file path
file_path = r"data.csv"

df = pd.read_csv(file_path)
```

```python
df.info()
```

```python
df.columns
```

```python
# List of columns to remove
columns_to_remove = [
    'goal', 'target', 'indicator', 'seriesCode', 'seriesDesc', 'release', 
    'type', 'unitsCode', 'unitsDesc', 'unitmultiplierCode', 
    'unitmultiplierDesc', 'reportingTypeCode', 'reportingTypeDesc', 
    'valueDetails', 'footnotes', 'sources', 'timeDetails', 'nature'
]

# Remove the columns
df = df.drop(columns=columns_to_remove)

# Display the updated DataFrame
df.head()
```

```python
df.columns
```

```python
# Calculate the number of null values for each column
null_values = df.isnull().sum()

# Calculate the total number of rows in the DataFrame
total_rows = len(df)

# Calculate the percentage of null values in each column
null_percentage = (null_values / total_rows) * 100

# Print the percentage of null values for each column
print("Percentage of null values in each column:")
print(null_percentage)
```

```python
# Replace null values with the mean in columns with more than 10% null values
columns_with_high_nulls = null_percentage[null_percentage > 10].index

for column in columns_with_high_nulls:
    if df[column].dtype in ['float64', 'int64']:  # Ensure the column is numeric
        mean_value = df[column].mean()
        df[column].fillna(mean_value, inplace=True)
    else:
        # Handle non-numeric columns if needed (e.g., filling with mode or a placeholder)
        mode_value = df[column].mode()[0] if not df[column].mode().empty else None
        if mode_value is not None:
            df[column].fillna(mode_value, inplace=True)

# Verify if the null values have been handled
null_values_after = df.isnull().sum()
print("Null values after handling:")
print(null_values_after)
```

```python
# Check total null values in each column
print(df.isnull().sum())
```

```python
# Find rows with any null values
rows_with_nulls = df[df.isnull().any(axis=1)]

# Display rows with null values
print(rows_with_nulls)
```

```python
import pandas as pd

# Function to fill NaN values with the row mean (from specific columns)
def fill_with_row_mean(row, columns):
    # Compute the mean of the row considering only non-NaN values from the selected columns
    row_values = row[columns]
    row_mean = row_values[row_values.notna()].mean()  # Only consider non-NaN values in the row

    # Fill NaN values in the specified columns with the calculated row mean
    # Update only NaN cells with the row mean, without changing other values
    row[columns] = row[columns].apply(lambda x: row_mean if pd.isna(x) else x)
    return row

# Columns to fill with row mean (must match your dataset's column names)
columns_to_fill = [
    'value_2000', 'value_2001', 'value_2002', 'value_2003', 'value_2004',
    'value_2005', 'value_2006', 'value_2007', 'value_2008', 'value_2009',
    'value_2010', 'value_2011', 'value_2012', 'value_2013', 'value_2014',
    'value_2015', 'value_2016', 'value_2017', 'value_latest_year', 'latest_year'
]

# Before filling, check the row values for any nulls
print("Before filling:")
print(df[columns_to_fill].iloc[0])  # Print first row's values in the relevant columns

# Apply the row mean filling function
df = df.apply(lambda row: fill_with_row_mean(row, columns_to_fill), axis=1)

# After filling, check the row values for any nulls
print("After filling:")
print(df[columns_to_fill].iloc[0])  # Print first row's values in the relevant columns

# Double check if any NaNs remain in the specified columns
print("NaN counts in the columns after filling:")
print(df[columns_to_fill].isnull().sum())  # Should print all zeros
```
